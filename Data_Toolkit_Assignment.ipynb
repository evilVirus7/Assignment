### 3. Explain the following terms: ###

**. The difference in np.array, np.asarray and np.asanyarray.**

**. The difference between Deep copy and shallow copy.** ###

Ans-

1. **np.array, np.asarray, and np.asanyarray**

These are three functions in NumPy used to create arrays. The key difference lies in their handling of input data types and the creation of new arrays.

a. **np.array:**

Creates a new array.

Converts input to a NumPy array.

Copies the input data if necessary.

Can handle various input types, including lists, tuples, and other NumPy arrays.

b. **np.asarray:**

Creates a new array if necessary.

If the input is already a NumPy array, it returns the input array itself without copying.

Primarily used for converting non-NumPy arrays to NumPy arrays efficiently.

c. **np.asanyarray:**

Similar to np.asarray but more flexible.

If the input is already an ndarray, it returns the input array without copying.

If the input is a subclass of ndarray, it returns the input array.

Otherwise, it creates a new ndarray.

2. **Deep Copy and Shallow Copy**

When copying arrays, it's important to understand the difference between deep and shallow copies:

a. **Shallow Copy:**

Creates a new array object but shares the same underlying data with the original array.

Changes made to one array will affect the other.

Use the = operator or np.array() to create a shallow copy.

b. **Deep Copy:**

Creates a new array object with a completely independent copy of the data.

Changes made to one array will not affect the other.

Use np.copy() or np.deepcopy() to create a deep copy.

```python
import numpy as np

# Shallow copy
arr1 = np.array([1, 2, 3])
arr2 = arr1
arr2[0] = 10
print(arr1)  # Output: [10, 2, 3]

# Deep copy
arr3 = np.copy(arr1)
arr3[0] = 20
print(arr1)  # Output: [10, 2, 3]
```

### 7. Clean and transform the 'Phone' column in the sample dataset to remove non-numeric characters and convert it to a numeric data type. Also display the table attributes and data types of each column. ###


```python
#Ans-

import pandas as pd

# Sample dataset
data = {'Name': ['Shelia', 'Sheryl', 'Sherry'],
        'Phone': ['857.139.8239', '(599)782-0605', '+91 8537800927']}
df = pd.DataFrame(data)

# Clean the 'Phone' column
def clean_phone_number(phone_number):
    """Cleans a phone number by removing non-numeric characters."""
    return ''.join(char for char in phone_number if char.isdigit())

df['Phone'] = df['Phone'].apply(clean_phone_number)
df['Phone'] = pd.to_numeric(df['Phone'], errors='coerce')

# Display table attributes and data types
print(df.info())
print(df)
```

c) **Label the x-axis as 'Value' and the y-axis as 'Frequency/Probability'.**

d) **Set the title of the plot as 'Histogram with PDF Overlay'.**

Ans - **Explanation:**

a. **Generate Data:**

np.random.randn(1000) generates 1000 random samples from a standard normal distribution.

b. **Plot Histogram:**

plt.hist() plots a histogram of the data with 30 bins.

density=True normalizes the histogram to a probability density function.

alpha=0.6 sets the transparency of the histogram bars.

c. **Plot PDF:**

np.linspace(-4, 4, 100) generates 100 evenly spaced points between -4 and 4.

norm.pdf(x) calculates the probability density function values for these points.

plt.plot() plots the PDF as a red line.

d. **Set Labels and Title:**

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import norm

# Create a NumPy array with 1000 samples from a normal distribution
data = np.random.randn(1000)

# Plot a histogram with 30 bins
plt.hist(data, bins=30, density=True, alpha=0.6, label='Histogram')

# Generate x-axis values for the PDF
x = np.linspace(-4, 4, 100)

# Plot the PDF
plt.plot(x, norm.pdf(x), 'r-', label='PDF')

# Set labels and title
plt.xlabel('Value')
plt.ylabel('Frequency/Probability')
plt.title('Histogram with PDF Overlay')

# Display legend
plt.legend()

# Show the plot
plt.show()
```

### 17. Create a Seaborn scatter plot of two random arrays, color points based on their position relative to the origin (quadrants), add a legend, label the axes, and set the title as 'Quadrant-wise Scatter Plot'. ###

Ans - **Explanation:**

a. **Generate Random Data:**

np.random.randn() generates 100 random numbers from a standard normal distribution for both x and y coordinates.

b. **Determine Quadrant:**

np.where() is used to determine the quadrant for each point based on its x and y coordinates.

c. **Create DataFrame:**

A pandas DataFrame is created to store the x, y, and quadrant information.

d. **Create Scatter Plot:**

sns.scatterplot() creates the scatter plot.

x and y specify the x and y coordinates.

hue='Quadrant' colors the points based on the quadrant.

data=df specifies the DataFrame to use.

e. **Set Labels and Title:**

plt.xlabel(), plt.ylabel(), and plt.title() set the axis labels and plot title.

f. **Show Plot:**

plt.show() displays the plot.

```python
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd

x = np.random.randn(100)
y = np.random.randn(100)

df = pd.DataFrame({'x': x, 'y': y})

def get_quadrant(row):
    if row['x'] > 0 and row['y'] > 0:
        return 'Positive X, Positive Y'
    elif row['x'] > 0 and row['y'] < 0:
        return 'Positive X, Negative Y'
    elif row['x'] < 0 and row['y'] > 0:
        return 'Negative X, Positive Y'
    else:
        return 'Negative X, Negative Y'

df['Quadrant'] = df.apply(get_quadrant, axis=1)

sns.scatterplot(x='x', y='y', hue='Quadrant', data=df)

plt.xlabel('X-axis')
plt.ylabel('Y-axis')
plt.title('Quadrant-wise Scatter Plot')

plt.show()
```

### 18. With Bokeh, plot a line chart of a sine wave function, add grid lines, label the axes, and set the title as 'Sine Wave Function'. ###

Ans - **Here's a breakdown of the code:**

a. **Import Bokeh:** Imports the necessary modules from the Bokeh library.

b. **Create a Figure:**

figure() creates a new figure object.

title, x_axis_label, and y_axis_label are used to set the title and axis labels.

c. **Generate Data:**

np.linspace() generates 100 evenly spaced points between 0 and 2π.

np.sin(x) calculates the sine of each point.

d. **Plot the Line:**

p.line() plots the line using the generated x and y values.

p.grid.grid_line_alpha = 0.3 sets the alpha value of the grid lines to 0.3, making them slightly transparent.

```python
from bokeh.plotting import figure, show

# Create a figure
p = figure(title="Sine Wave Function", x_axis_label='x', y_axis_label='y')

# Generate x and y values for the sine wave
x = np.linspace(0, 2*np.pi, 100)
y = np.sin(x)

# Plot the line
p.line(x, y, line_width=2)

# Add grid lines
p.grid.grid_line_alpha = 0.3

# Show the plot
show(p)
```

```python
from bokeh.plotting import figure, show
from bokeh.models import HoverTool
import numpy as np

# Generate random categorical data and values
categories = ['A', 'B', 'C', 'D', 'E']
values = np.random.randint(1, 20, 5)

# Create a figure
p = figure(x_range=categories, title="Random Categorical Bar Chart",
           toolbar_location=None, tools="")

# Add a bar chart
p.vbar(x=categories, top=values, width=0.5, color=['red', 'green', 'blue', 'orange', 'purple'])

# Add hover tooltips
hover = HoverTool(tooltips=[("Category", "@x"), ("Value", "@top")])
p.add_tools(hover)

# Show the plot
show(p)
```
